## Rata-rata Konsumsi Daging Dunia
    Dari data yang diperoleh dari data.oecd.org, kita mengambil rata-rata konsumsi daging dunia lalu di-plot.

### WORLD - Plot - Konsumsi Daging

```python
world = pd.read_csv("country/world.csv", index_col="TIME")

world_meat=pd.DataFrame(world['VALUE'][:29])

for i in range(len(world_meat)):
    world_meat['VALUE'][1990+i]=np.sum(world["VALUE"][1990+i])

plt.plot(world_meat["VALUE"],label="Konsumsi Daging Dunia")
plt.xlabel("Tahun")
plt.ylabel("Konsumsi Daging (Kg per Kapita)")
plt.legend()
```

### WORLD - Plot - GDP

```python
world_gdp = gdp['WLD'] 
plt.plot(world_gdp,label="GDP Dunia")
plt.xlabel("Tahun")
plt.ylabel("GDP (Triliun USD)")
plt.legend()
```

### WORLD - Plot - Perbandingan Perkembangan Data

```python
world_meat_dif=pd.DataFrame(world['VALUE'][:28])

for i in range(1,len(world_meat)):
    world_meat_dif["VALUE"][i+1990-1]=((world_meat["VALUE"][i+1990]-world_meat["VALUE"][i+1990-1])/world_meat["VALUE"][i+1990-1]*100)

world_gdp_dif=pd.DataFrame(world['VALUE'][:28])
               
for i in range(1,len(world_gdp)):
    world_gdp_dif["VALUE"][i+1990-1]=((world_gdp[i]-world_gdp[i-1])/world_gdp[i-1]*100)

plt.plot(world_meat_dif, label="Konsumsi Daging")
plt.plot(world_gdp_dif, label="GDP")
plt.ylabel("Selisih (%)")
plt.xlabel("Tahun")
plt.legend()
```

### Konsumsi Daging pada beberapa Negara Berkembang di Asia Tenggara
    Negara yang akan disandingkan disini meliputi Indonesia, Malaysia, Vietnam, Thailand, dan Filipina. Kelima negara ini   sama-sama negara berkembang dengan kondisi geografis berdekatan sehingga kurang lebih sama. (Asia Tenggara)

```python
indonesia = pd.read_csv("country/indonesiA.csv", index_col="TIME")
malaysia = pd.read_csv("country/malaysia.csv", index_col="TIME")
vietnam = pd.read_csv("country/vietnam.csv", index_col="TIME")
thailand = pd.read_csv("country/thailand.csv", index_col="TIME")
philippines = pd.read_csv("country/philippines.csv", index_col="TIME")
```

#### SEA - Plot - Total Konsumsi Daging Gabungan

```python
id_meat=pd.DataFrame(indonesia['VALUE'][:29])
my_meat=pd.DataFrame(malaysia['VALUE'][:29])
vn_meat=pd.DataFrame(vietnam['VALUE'][:29])
th_meat=pd.DataFrame(thailand['VALUE'][:29])
ph_meat=pd.DataFrame(philippines['VALUE'][:29])

for i in range(len(id_meat)):
    id_meat['VALUE'][1990+i]=np.sum(indonesia["VALUE"][1990+i])
    my_meat['VALUE'][1990+i]=np.sum(malaysia['VALUE'][1990+i])
    vn_meat['VALUE'][1990+i]=np.sum(vietnam['VALUE'][1990+i])
    th_meat['VALUE'][1990+i]=np.sum(thailand['VALUE'][1990+i])
    ph_meat['VALUE'][1990+i]=np.sum(philippines['VALUE'][1990+i])

plt.plot(id_meat, label ="Indonesia")
plt.plot(my_meat, label ="Malaysia")
plt.plot(vn_meat, label ="Vietnam")
plt.plot(th_meat, label ="Thailand")
plt.plot(ph_meat, label ="Filipina")
plt.xlabel("Tahun")
plt.ylabel("Konsumsi Daging (Kg per Kapita)")
plt.legend()
```

Dimulai dari tingkat konsumsi daging secara keseluruhan, kita langsung dapat melihat trend yang sangat meningkat dari Vietnam, meskipun belum dapat menyentuh Malaysia. Filipina juga mengalami trend naik namun belum berada di level dua negara tadi. Indonesia dan Thailand hanya mengalami naik-turun yang tidak begitu signifikan, dan hanya berada di tempat terbawah dan kedua terbawah.
  
  Lalu apa yang menyebabkan turun dan naiknya tingkat konsumsi daging ini? Apakah daya beli masyarakat dan tingkat ekonomi suatu negara dapat mempengaruhinya? Selanjutnya kita akan mencoba melihat pengaruh GDP.

#### SEA - Plot - GDP

```python
indonesiagdp = gdp['IDN']
malaysiagdp = gdp["MYS"]
vietnamgdp = gdp["VNM"]
thailandgdp = gdp["THA"]
philippinesgdp = gdp["PHL"]
plt.plot(indonesiagdp,label="Indonesia")
plt.plot(malaysiagdp,label="Malaysia")
plt.plot(vietnamgdp,label="Vietnam")
plt.plot(thailandgdp,label="Thailand")
plt.plot(philippinesgdp,label="Philippines")
plt.xlabel("Tahun")
plt.ylabel("GDP (Triliun USD)")
plt.legend()
```

Jika dibandingkan dengan plot sebelumnya dimana 2 terendah tingkat konsumsi daging ditempati Indonesia dan Thailand, Ternyata plot tingkat GDP menunjukkan hasil yang(sort of)sebaliknya! **Indonesia memiliki GDP tertinggi** dengan trend yang terus naik meskipun sempat "drop" pada tahun 1998, dan Thailand menyusul di tempat kedua. Vietnam disini justru memiliki GDP paling rendah, dengan Filipina dan Malaysia diatasnya. 

Apakah hal ini berarti nilai GDP tidak berpengaruh sama sekali pada kenaikan tingkat konsumsi daging?<br>

Sebenarnya hal itu kurang tepat. Penjelasan yang masuk akal untuk Indonesia yang GDPnya paling tinggi namun Konsumsi Dagingnya paling rendah adalah karena **Indonesia dikenal sebagai negara maritim dan negara agraris**. Jadi konsumsi indonesia lebih ke ranah lain. Faktor keagamaan juga berpengaruh karena daging babi ikut menjadi indikator, sedangkan Mayoritas Penduduk Indonesia adalah Muslim yang **tidak memakan daging babi**.<br> 
<br>
Lebih dalam mengenai pengaruh GDP, selanjutnya kita akan melihat perbandingannya dari Persentase Kenaikan atau Penurunan.
